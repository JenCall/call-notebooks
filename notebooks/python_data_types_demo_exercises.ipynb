## 1. What are Data Types?

A **data type** tells Python what kind of value it is working with.

Examples:
- `5` is an integer
- `3.14` is a float
- `"hello"` is a string
- `True` is a boolean

The built-in function `type()` shows the type of a value.


## 2. Integers (`int`)

Integers are **whole numbers**: positive, negative, or zero.

Examples:
- `8`
- `-3`
- `0`

Common operations:
- addition: `+`
- subtraction: `-`
- multiplication: `*`
- division: `/`
- remainder (modulo): `%`
- comparison: `>`, `<`, `==`


```python
print("Examples of integers:")
print(8)
print(-3)
print(0)

print("\nType of 8:")
print(type(8))
```

```python
print("Addition:", 3 + 5)
print("Subtraction:", 5 - 8)
print("Multiplication:", 3 * 8)
print("Division:", 8 / 2)
print("Modulo (remainder):", 10 % 3)
print("Comparison:", 3 > 2)
```

### Why modulo matters

The operator `%` gives the **remainder after division**.

Examples:
- `10 % 3` is `1`
- `8 % 2` is `0`

It is useful for checking if a number is **even or odd**.


```python
numbers = [7, 8, 9, 10]

for n in numbers:
    if n % 2 == 0:
        print(n, "is even")
    else:
        print(n, "is odd")
```

## 3. Floating-Point Numbers (`float`)

Floats represent **real numbers with decimal points**.

Examples:
- `3.14`
- `-0.5`
- `2.0`

Floats can do operations similar to integers.


```python
print("Examples of floats:")
print(3.14)
print(-0.5)
print(2.0)

print("\nType of 3.14:")
print(type(3.14))
```

```python
print("Addition:", 3.5 + 2.1)
print("Division:", 7 / 2)
print("Comparison:", 3.2 > 1.8)
```

### Floating-point precision

Sometimes floats produce small rounding differences because they are stored in **binary**.


```python
print(0.1 + 0.2)
print(round(0.1 + 0.2, 1))
```
