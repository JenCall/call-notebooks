### 1. Elif Chains vs Nesting

```python
# Elif chain (mutually exclusive)
grade = 85
if grade >= 90:
    print("A")
elif grade >= 80:
    print("B")
elif grade >= 70:
    print("C")
else:
    print("F")

# Nesting (when hierarchy matters)
user_logged_in = True
user_is_admin = False

if user_logged_in:
    if user_is_admin:
        print("Welcome, Admin")
    else:
        print("Welcome, Guest User")
else:
    print("Please log in first")
```

### 2. Guard Clauses

```python
# guard clause: handle edge case early
def divide(a, b):
    if b == 0:
        return "Error: division by zero"
    return a / b

print(divide(10, 2))   # 5.0
print(divide(10, 0))   # Error
```

### 3. Truthiness in Python

```python
items = []
if items:   # Better
    print("We have items")
else:
    print("Empty list")

items = [1, 2, 3]
if len(items) > 0:  # Explicit, but verbose
    print("We have items")
```

### 4. Logical Operators

```python
user_name = "Anna"
professor = True
if user_name == "Anna" and professor:
  print(f"Welcome, {user_name}.")
```

### 5. Value Comparisons

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)   # True (value equality)
print(a is b)   # False (different objects)

# case-insensitive match
command = "Exit"
if command.lower() == "exit":
    print("Goodbye")
```

### 6. For Loop

```python
names = ["Alice", "Bob", "Charlie"]
for name in names:
    print(f"Hello, {name}!")
```

### 7. Iterating with range

```python
n = 5
total = 0
for i in range(n + 1):
    total += i
print("Sum of first", n, "numbers:", total)
```

### 8. Iterating over Sequences

```python
person = {"name": "Alice", "age": 25}
for key, value in person.items():
    print(f"{key}: {value}")
```

### 9. Enumerate

```python
lines = ["first", "second", "third"]
for index, text in enumerate(lines, start=1):
    print(f"{index}: {text}")
```

### 10. Comprehensions

```python
squares = [x*x for x in range(10) if x % 2 == 0]
print(squares)
```

### 11. Break & Continue

```python
numbers = [1, 3, 5, 8, 9]
for n in numbers:
    if n % 2 == 0:
        print("First even:", n)
        break
```

### 12. Else on Loops

```python
numbers = [1, 3, 5, 7]
for n in numbers:
    if n % 2 == 0:
        print("Even found:", n)
        break
else:
    print("No even numbers found")
```

### 13. While Loop Essentials

```python
n = 5
while n >= 0:
    print(n)
    n -= 1
print("Liftoff!")
```
