How would you create two NumPy arrays of random numbers with a million elements each?

```python
n= 1000000
a = np.random.rand(n)
b = np.random.rand(n)
```

What does this code define?

```python
import time

class Timer:
    def __init__(self):
        self.times = []

    def start(self):
        self.tik = time.time()

    def stop(self):
        self.times.append(time.time() - self.tik)
        return self.times[-1]
```

How is the `Timer` class used to measure the time taken for element-wise addition using a loop and using NumPy's vectorized addition?

```python
timer = Timer()

timer.start()
c = np.zeros(n)
for i in range(n):
    c[i] = a[i] + b[i]
timer.stop()
```

```python
timer.start()
c = a + b
timer.stop()
```

What are the recorded times for the two addition methods?

```python
timer.times
```

What is the speedup of the NumPy vectorized addition compared to the loop?

```python
print("speedup is: %s" % (timer.times[0]/timer.times[1]))
```
